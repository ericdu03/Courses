```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def expansion(nmax):
    cn = []
    count = 0
    for n in range(1, nmax+1):
        if n%2 == 0: 
            cn.append(0)
        else:
            cn.append(2 * np.sqrt(2) * (-1)**count /((2 * count + 1)**2 * np.pi**2))
            count+=1 
    fn = lambda x: np.sum([ cn[n] * np.sqrt(2) * np.sin((n+1) * np.pi * x) for n in range(len(cn))])
    return fn

partial1 = expansion(3)
partial3 = expansion(5)
partial10 = expansion(10)
partial50 = expansion(50)

x_values = np.linspace(0, 1, 200)

def triangle(x):
    if 0 <= x <= 0.5:
        return x
    else: 
        return 1 - x
    
v_triangle = np.vectorize(triangle)
true_triangle = [v_triangle(x_values[i]) for i in range(len(x_values))]

plt.scatter(x_values, [partial1(x_values[i]) for i in range(len(x_values))], s = 0.4)
plt.scatter(x_values, [partial3(x_values[i]) for i in range(len(x_values))], s = 0.4)
plt.scatter(x_values, [partial10(x_values[i]) for i in range(len(x_values))], s = 0.4)
plt.scatter(x_values, [partial50(x_values[i]) for i in range(len(x_values))], s = 0.4)
plt.plot(x_values, true_triangle)
plt.legend(['true curve', 'n = 3', 'n = 5', 'n = 10', 'n = 50'])
plt.show()
```
