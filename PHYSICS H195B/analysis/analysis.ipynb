```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import glob
```

```python
for file in glob.glob("data/long-*.csv"):
    data = pd.read_csv(file, names = ["f", "VSWR1", "VSWR2"], skiprows = 1)
    # plt.figure()
    plt.plot(data["f"], data["VSWR1"], label = f"{file[5:]} - VSWR1")
    # plt.plot(data["f"], data["VSWR2"])
    plt.legend()
    # plt.legend(["VSWR1", "VSWR2"])
    plt.title("Long waveguides (20cm)")

    plt.xlabel("Frequency (GHz)")
    plt.ylabel("VSWR")
```

```python
file = "data/long-finitecond-bend-round-big.csv"
data = pd.read_csv(file, names = ["f", "VSWR1", "VSWR2"], skiprows = 1)
# plt.figure()
plt.plot(data["f"], data["VSWR1"], label = f"VSWR")
# plt.plot(data["f"], data["VSWR2"])
plt.legend()
# plt.legend(["VSWR1", "VSWR2"])
plt.title("20cm waveguide, large rounded bend")

plt.xlabel("Frequency (GHz)")
plt.ylabel("VSWR")
plt.show()
```
