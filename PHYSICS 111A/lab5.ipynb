```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt
```

```python
Vgs = np.arange(0, -1.9, -0.1)
Id = np.array([2.04, 1.88, 1.72, 1.57, 1.42, 1.28, 1.14, 1, 0.87, 0.75, 0.63, 0.51, 0.41, 0.32, 0.23, 0.15, 0.09, 0.04, 0.01])

plt.scatter(Vgs, Id, s = 0.5, c = 'blue')
plt.title("Vgs vs ID")
plt.xlabel("Vgs")
plt.ylabel("Id")
plt.show()
```

```python
def fit(Vgs, Vp, IDSS):
    return IDSS*(1 - (Vgs/Vp))**2


func, cov = opt.curve_fit(fit, Id, Vgs)
IDSS_fit, Vp_fit = func

plt.scatter(Vgs, Id, s = 0.8, c = 'blue')
plt.plot(Vgs, fit(Vgs, Vp_fit, IDSS_fit), c = 'orange')
plt.show()

print(IDSS_fit, Vp_fit)
# plt.scatter(Vgs, Id/1000, s = 0.8, c = 'blue')
# plt.plot(Vgs, -Vgs/510, c = 'green')
# plt.plot(Vgs, -Vgs/1000, c = 'purple')
# plt.plot(Vgs, -Vgs/2000, c = 'red')
# plt.legend(["Data", "510 load line", "1k load line", "2k load line"])
# plt.title("Vgs vs ID")
# plt.xlabel("Vgs")
# plt.ylabel("Id")
# plt.show()
# gm = IDSS_fit * 2 * (1 - Vgs/Vp_fit) * -1/Vp_fit
# plt.scatter(Vgs, gm)
# plt.show()
```

```python
gm = -2 * IDSS_fit/1000 * (1 - Vgs/Vp_fit) * 1/Vp_fit

gm = gm[:-2]
plt.scatter(Vgs[:-2], 1/gm, s = 0.8)
plt.ylabel("1/gm")
plt.xlabel("Vgs")
plt.show()
```
