```python
import pandas as pd
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Problem_4.2_Data.csv", skiprows=np.arange(0, 19, dtype=int), dtype=float)

V_R = data['Channel 1 (V)']
V_D = data['Channel 2 (V)']

I_R = V_R[100:-100]/1000
V_D = V_D[100:-100]
I_D = I_R.copy()

slice1 = V_D > 0.3
I_R = I_R[slice1]
I_D = I_D[slice1]
V_D = V_D[slice1]


plt.plot(V_D, np.log(I_D))
plt.title("V_D vs. log(I_D), first diode")
plt.show()

def linear(x, m, b):
    return m * x + b

vals, cov = opt.curve_fit(linear, V_D, np.log(I_D))
slope, y_int = vals

plt.plot(V_D, np.log(I_D))
plt.plot(V_D, slope *V_D + y_int)
plt.title("Theoretical vs. Experimental Plots")
plt.show()


print((1/slope)/(25e-3), np.exp(y_int))
print(y_int)

print("Experimental:")
print(np.average(I_D[(V_D > 0.39)& (V_D < 0.41)]))
print(np.average(I_D[(V_D > 0.49)& (V_D < 0.51)]))
print(np.average(I_D[(V_D > 0.59)& (V_D < 0.61)]))
print("\n")
x_values = np.linspace(0.3, 0.65, 1000)
y_values = slope * x_values + y_int

print("Theoretical:")
print(np.exp(np.average(y_values[(x_values > 0.39)& (x_values < 0.41)])))
print(np.exp(np.average(y_values[(x_values > 0.49)& (x_values < 0.51)])))
print(np.exp(np.average(y_values[(x_values > 0.59)& (x_values < 0.61)])))
```
