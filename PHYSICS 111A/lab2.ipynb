```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("Problem_2.2_Data.csv", skiprows=np.arange(0, 20, dtype=int), dtype=float)

f = np.linspace(1000, 10e5, 20000)
R = 3000 # resistor
c = 4.7e-9 # capacitor
w = 2 * np.pi * f
gain = R*w* c/np.sqrt(1 + (R * w * c)**2)

j = 1j
H = R*j*w*c/(1 + R * j * w * c)
phase = np.arctan(np.imag(H)/np.real(H)) * 180/np.pi

fig, (ax1, ax2) = plt.subplots(2, 1)
ax1.plot(data['Frequency (Hz)'], data['Channel 1 Magnitude (X)'], c = 'orange')
ax1.plot(data['Frequency (Hz)'], data['Channel 2 Magnitude (X)'], c = 'b')
ax1.plot(f, gain, c = 'green')
ax1.legend(['Channel 1', 'Channel 2', 'Theoretical'])
ax1.set_xscale('log')

ax2.plot(data['Frequency (Hz)'], data['Channel 2 Phase (deg)'], c = 'b')
ax2.plot(f, phase, c = 'green')
ax2.set_xscale('log')
ax2.legend(['Measured', 'Theoretical'])
plt.show()  

print(phase)
# print(data['Channel 2 Phase (deg)'])
```
