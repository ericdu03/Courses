```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("Problem_2.2_Data.csv", skiprows=np.arange(0, 20, dtype=int), dtype=float)

fig, (ax1, ax2) = plt.subplots(2, 1)
ax1.plot(data['Frequency (Hz)'], data['Channel 1 Magnitude (X)'], c = 'orange')
ax1.plot(data['Frequency (Hz)'], data['Channel 2 Magnitude (X)'], c = 'b')
ax1.legend(['Channel 1', 'Channel 2'])
ax1.set_xscale('log')

ax2.plot(data['Frequency (Hz)'], data['Channel 2 Phase (deg)'], c = 'b')
ax2.set_xscale('log')
plt.show()
```
