```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt
from scipy.interpolate import interp1d
np.set_printoptions(threshold=np.inf)


intensity = np.linspace(100, 0, 6)/100
current = [0.556, 0.445, 0.333, 0.222, 0.111, 0]

plt.scatter(intensity, current)
plt.xlabel("Intensity")
plt.ylabel("Measured current")
plt.title("Intensity vs. Current")
plt.show()
```

```python
def linear(x, m): 
    return m*x


optimized, cov = opt.curve_fit(linear, intensity, current)
m  = optimized

theoretical = [m * intensity[i] for i in range(len(intensity))]

plt.plot(intensity, theoretical, label ="Fitted Data", c = 'r')
plt.scatter(intensity, current, c = 'b', label = 'Simulated data')
plt.xlabel("Intensity")
plt.ylabel("Current")
plt.title("Intensity vs. Current")
plt.legend() 
plt.show()
```
